```
import matplotlib.pyplot as plt
```

```
from PIL import Image
nb_image = 8
images = []
size = 640,480
for i in range(1,nb_image+1):
  imname = "./BE_groupe_4/images/Nadar/Nadar"+str(i)
  im = Image.open(imname+".PNG")
  im_resized = im.resize(size, Image.ANTIALIAS)
  im_resized = im_resized.convert('RGB')
  im_resized.save(imname+"_resized.jpg", "JPEG")
  images.append(im_resized)
plt.figure(1)
plt.imshow(images[1])
plt.show()
```
